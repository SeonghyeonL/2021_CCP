```python
import pandas as pd
import numpy as np
!pip install pyproj
from pyproj import Proj, transform
import folium

# Pandas Float 자릿수 표시 제한
#pd.options.display.float_format = '{:.6f}'.format
```

```python
# Projection 정의
# 중부원점(Bessel): 서울 등 중부지역 EPSG:2097
proj_1 = Proj(init='epsg:5186')

# WGS84 경위도: GPS가 사용하는 좌표계 EPSG:4326
proj_2 = Proj(init='epsg:4326')
```

```python
# 데이터 프레임 지정
#temp = d_1905[:50].copy()
#df = d_1905[:50].copy()

#x_list = []
#y_list = []

#for idx, row in temp.iterrows():
#    x, y = row['x'], row['y']
#    x_, y_ = transform(proj_1, proj_2, x, y)
#    x_list.append(x_)
#    y_list.append(y_)
    
#df['long'] = x_list # 경도
#df['lati'] = y_list # 위도

# 시간 단축
df2 = d_1905[:200].copy()
converted = transform(proj_1, proj_2, df2['x'].values, df2['y'].values)
df2['long'] = converted[0]
df2['lati'] = converted[1]
#df2['long'] = converted[0] - 1.2665
#df2['lati'] = converted[1] + 0.0055
```

```python
df2[-10:]
```

```python
# 시각화

m = folium.Map(location=[37.583772, 127.037004], zoom_start=8)
#from folium.plugins import MarkerCluster
#marker_cluster = MarkerCluster().add_to(m)

for i in range(len(df2)):
    lati_ = df2['lati'][i]
    long_ = df2['long'][i]
    folium.Marker([lati_, long_],tooltip=df2['id'][i]).add_to(m)
```
