```python
import pandas as pd
```

```python
data1 = pd.read_csv('/content/drive/Shareddrives/2021 CCP/천신만고 끝에 얻은 데이터/KT 데이터 재구성/extend_ver/202107KT_ymdmf.csv')
data1 = data1.drop(['Unnamed: 0'], axis=1)
data1
```

```python
data = data1[['year','month','day','timezn_cd','week','weekday','holiday','id','x','y','admi_cd','total','male','female']]
```

```python
data['young'] = data1['m20'] + data1['m25'] + data1['f20'] + data1['f25']
data['middle'] = data1['m30'] + data1['m35'] + data1['m40'] + data1['m45'] + data1['f30'] + data1['f35'] + data1['f40'] + data1['f45']
data['older'] = data1['m50'] + data1['m55'] + data1['m60'] + data1['f50'] + data1['f55'] + data1['f60']
data['oldest'] = data1['m65'] + data1['m70'] + data1['m75'] + data1['m80'] + data1['m85'] + data1['f65'] + data1['f70'] + data1['f75'] + data1['f80'] + data1['f85']
```

```python
weather = pd.read_csv('/content/drive/Shareddrives/2021 CCP/천신만고 끝에 얻은 데이터/2022.01.28/2021.07.tr.csv')
corona = pd.read_excel('/content/drive/Shareddrives/2021 CCP/천신만고 끝에 얻은 데이터/2022.01.28/2021.07.xlsx')
```

```python
data['temperature'] = 0
data['rain'] = 0
data['corona'] = 0
```

```python
for i in range(len(weather)):
  data.loc[data['day'] == int(weather['date'][i][7:]), ['temperature','rain']] = [weather['temperature'][i],weather['rain'][i]]
```

```python
for i in range(len(corona)):
  data.loc[data['day'] == corona['day'][i], ['corona']] = corona['confirmed'][i]
```

```python
data.to_csv("/content/drive/Shareddrives/2021 CCP/천신만고 끝에 얻은 데이터/2022.01.28/202107_add.csv", mode='w', index = None)
```
