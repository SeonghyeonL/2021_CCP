```python
# 2019년 5월 기준
# 일요일: 5, 12, 19, 26 -> 7로 나눈 나머지가 6
# 공휴일: 5(어린이날), 6(대체공휴일), 12(석가탄신일)

# 2019년 7월 기준
# 일요일: 7, 14, 21, 28 -> 7로 나눈 나머지가 5

# 2020년 5월 기준
# 일요일: 3, 10, 17, 24, 31 -> 7로 나눈 나머지가 1
# 공휴일: 5(어린이날)

# 2020년 7월 기준
# 일요일: 5, 12, 19, 26 -> 7로 나눈 나머지가 0

# 2021년 5월 기준
# 일요일: 2, 9, 16, 23, 30 -> 7로 나눈 나머지가 4
# 공휴일: 5(어린이날), 19(석가탄신일)

# 2021년 7월 기준
# 일요일: 4, 11, 18, 25 -> 7로 나눈 나머지가 3

# week   : 일(0) 월(1) 화(2) 수(3) 목(4) 금(5) 토(6)
# weekday: 일(0) 월(1) 화(1) 수(1) 목(1) 금(1) 토(0) → 워드 내용 변경함!
# holiday: 공휴일(1) 공휴일아님(0)
```

```python
def func_week(par):  # 무슨 요일인가
  if par//100 == 201905: return (par+1)%7
  elif par//100 == 201907: return (par+2)%7
  elif par//100 == 202005: return (par+6)%7
  elif par//100 == 202007: return (par+0)%7
  elif par//100 == 202105: return (par+3)%7
  elif par//100 == 202107: return (par+4)%7
```

```python
def func_weekday(par):  # 평일인가
  res = func_week(par)
  if res == 0 or res == 6: return 0  # 주말
  else: return 1  # 평일
```

```python
def func_holiday(par):  # 공휴일인가
  holi_list = []
  if par//100 == 201905:
    holi_list = [20190505, 20190506, 20190512]
  elif par//100 == 202005:
    holi_list = [20200505]
  elif par//100 == 202105:
    holi_list = [20210505, 20210519]

  if par in holi_list: return 1  # 공휴일
  else: return 0  # 공휴일아님
```

```python
d_1905['year'] = d_1905['etl_ymd'] // 10000
d_1905['month'] = d_1905['etl_ymd'] // 100 % 100
d_1905['day'] = d_1905['etl_ymd'] % 100
d_1905['week'] = 0     # default
d_1905['weekday'] = 0  # default
d_1905['holiday'] = 0  # default
for itr in range(len(d_1905)):
  d_1905['week'][itr] = func_week(d_1905['etl_ymd'][itr])
  d_1905['weekday'][itr] = func_weekday(d_1905['etl_ymd'][itr])
  d_1905['holiday'][itr] = func_holiday(d_1905['etl_ymd'][itr])
```

```python
temper = pd.read_csv('/content/drive/Shareddrives/2021 CCP/천신만고 끝에 얻은 데이터/KT 데이터 재구성/컬럼생성용데이터/제기동_3시간기온_201904_201905.csv')

temper.columns = ['day', 'hour', 'forecast', 'value']

# day 1, hour 200, forecast 4 → 한국 기준 1일 15시(=2+4+9)
# 15 18 21 0 3 6 9 12
temper = temper[temper['forecast'] == 4]
temper['hour'] = temper['hour'] / 100 + temper['forecast'] + 9
temper = temper[['day','hour','value']]
temper = temper[29*8+3:(30+30)*8+4] # 변경할 부분
temper = temper.reset_index(drop=True)
temper['day'] = pd.to_numeric(temper['day'])
temper.loc[0:4, 'day'] = 0

temper.loc[temper['hour'] > 23, 'day'] += 1
temper.loc[temper['hour'] > 23, 'hour'] -= 24
```

```python
# 빈 시간 채우기
f_temper = temper.copy()
f_temper['hour'] = (f_temper['day'] - 1) * 24 + f_temper['hour']
f_temper = f_temper[['hour','value']]
r_temper = f_temper.copy() # for temp
for i in range(len(f_temper)-1):
  b_hour = f_temper['hour'][i]
  b_value = f_temper['value'][i]
  a_hour = f_temper['hour'][i+1]
  a_value = f_temper['value'][i+1]
  r_temper = r_temper.append({'hour':b_hour+1, 'value':b_value+(a_value-b_value)/3*1}, ignore_index=True)
  r_temper = r_temper.append({'hour':b_hour+2, 'value':b_value+(a_value-b_value)/3*2}, ignore_index=True)
r_temper = r_temper.sort_values(by='hour')
r_temper = r_temper.reset_index(drop=True)
```

```python
f_temper = r_temper.copy()
f_temper['day'] = 1 + f_temper['hour'] // 24
f_temper['hour'] = f_temper['hour'] % 24
f_temper = f_temper[['day', 'hour', 'value']]
f_temper = f_temper.drop(len(f_temper)-1)
```
