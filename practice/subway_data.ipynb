```python
import pandas as pd
subdata = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/2021CCP/subwayyy.csv')
subdata = subdata.reset_index()
subdata
```

```python
subdata.columns = ['사용일자','노선명', '역명','승차총승객수','하차총승객수','e','f']
subdata = subdata.drop(['e','f'],axis=1)
subdata
```

```python
is_usable1 = subdata['노선명'] == '1호선'
is_usable2 = subdata['노선명'] == '2호선'
is_usable3 = subdata['노선명'] == '3호선'
is_usable4 = subdata['노선명'] == '4호선'
is_usable5 = subdata['노선명'] == '5호선'
is_usable6 = subdata['노선명'] == '6호선'
is_usable7 = subdata['노선명'] == '7호선'
is_usable8 = subdata['노선명'] == '8호선'
is_usable9 = subdata['노선명'] == '9호선'
is_usable = is_usable1|is_usable2|is_usable3|is_usable4|is_usable5|is_usable6|is_usable7|is_usable8|is_usable9

subdata = subdata[is_usable]
subdata = subdata.sort_values(by=['역명','사용일자'])
subdata = subdata.drop(['노선명'],axis=1)
subdata
```

```python
stations = subdata['역명'].values
stations = set(stations)
stations = list(stations)
stations.sort()
len(stations)
```

```python
subdata1 = subdata
for i in range(268):
  for j in range(20210501,20210532):
    date1 = subdata1['사용일자'] == j
    name1 = subdata1['역명'] == stations[i]
    d_and_n = date1&name1
    if sum(d_and_n)>1:
      tempdata = subdata1[d_and_n]
      subdata1 = subdata1[~d_and_n]
      t_data = {'사용일자':j,'역명':stations[i],'승차총승객수':tempdata.sum()[2],'하차총승객수':tempdata.sum()[3]}
      subdata1 = subdata1.append(t_data, ignore_index=True)
subdata1 = subdata1.sort_values(by=['역명','사용일자'])
subdata1
```

```python
testdata = subdata1[subdata1['사용일자']==20210531]
testdata
```

```python
testdata = testdata.drop(['사용일자'],axis=1)
testdata = testdata.set_index('역명')
testdata
```

```python
traindata = subdata1[subdata1['사용일자']!=20210531]
traindata
```

```python
traindata.mean()
```

```python
is_station = traindata['역명'] == stations[0]
train_mean = traindata[is_station]
train_mean.mean()
```

```python
from pandas import Series, DataFrame

s_m_res = {}

for i in range(268):
  is_station = traindata['역명'] == stations[i]
  train_mean = traindata[is_station]
  s_m_res[stations[i]] = [train_mean.mean()[1], train_mean.mean()[2]]

mean_res = DataFrame(s_m_res)
mean_res
```

```python
mean_res = mean_res.rename(index={0:'승차총승객수', 1:'하차총승객수'})
mean_res = mean_res.transpose()
mean_res
```

```python
total_res = mean_res - testdata
total_res
```

```python
total_res_1 = total_res['승차총승객수'].values.tolist()
total_res_2 = total_res['하차총승객수'].values.tolist()
len(total_res_2)
```

```python
res = []
for i in range(268):
  value = total_res_2[i] - total_res_1[i]
  if value == 0:
    res.append('보통')
  elif value > 0:
    res.append('혼잡')
  elif value < 0:
    res.append('여유')

res
```
